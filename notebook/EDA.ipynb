```python
import pandas as pd
import numpy as np
```

```python
data_file = {
    'riders': ('signup_date',),
    'sessions': ('session_time',),
    'trips': ('pickup_time', 'dropoff_time'),
    'promotions': ('start_date', 'end_date'),
    'drivers': ('signup_date','last_active')

}
dfs = {}
for name, date_cols in data_file.items():
    dfs[name] = pd.read_csv(f'../data/{name}.csv', parse_dates=list(date_cols))
```

```python
#checking data quality of the dataframes

def data_quality_check(df, name):
    print(f"{name.upper()}")
    print(f"{'-' * 60}")

    print(f"Duplicate Rows: {df.duplicated().sum()}")
    print(f"\nNull counts: )")
    
    print(df.isnull().sum())
    print(df.dtypes)
    print()

data = { 'sessions', 'trips', 'promotions', 'drivers', 'riders'}
for name in data:
    data_quality_check(dfs[name], name)

```
